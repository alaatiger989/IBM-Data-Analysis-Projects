```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Data
years = np.arange(1980 , 2014)
immigrants = np.random.randint(2000 , 10000 , size = (34 , ))
```

```python
fig , ax = plt.subplots()
ax.plot(years , immigrants)
plt.show()
```

```python
fig , ax = plt.subplots()
ax.scatter(years , immigrants)
plt.show()
```

```python
# limits on x-axis
plt.xlim(1975 , 2015)
```

```python
# Enabling Grid
plt.grid(True)
```

```python
ax.grid()
```

```python
# legend
plt.legend(["immigrants"])
```
